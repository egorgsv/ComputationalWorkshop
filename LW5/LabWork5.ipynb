```python
import numpy as np
import pandas as pd
```

```python
#Степенной метод
def power_method(A, eps):
    x_0 = np.array(np.ones(A.shape[0]))
    lamb = 0
    iter = 0
    while True:
        iter += 1
        x_1 = np.dot(A, x_0)
        if abs(x_1[0] / x_0[0] - lamb) < eps:
            lamb = x_1[0] / x_0[0]
            break
        lamb = x_1[0] / x_0[0]
        x_0 = x_1
    return lamb, iter
```

```python
def scal_method(A, eps):
    x_0 = np.array(np.ones(A.shape[0]))
    y_0 = np.array(np.ones(A.shape[0]))
    lamb = 0
    iter = 0
    while True:
        iter += 1
        x_1 = np.dot(A, x_0)
        y_1 = np.dot(A.T, y_0)
        if abs(np.dot(x_1, y_1) / np.dot(x_0, y_1) - lamb) < eps:
            lamb = np.dot(x_1, y_1) / np.dot(x_0, y_1)
            break
        lamb = np.dot(x_1, y_1) / np.dot(x_0, y_1)
        x_0 = x_1
        y_0 = y_1
    return lamb, iter
```

```python
def hilbert_matrix(n: int):
    matrix = np.zeros((n, n))
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            matrix[i - 1][j - 1] = 1 / (i + j - 1)
    return matrix
```

```python
result = []
for size in [3, 4, 5]:
    A = hilbert_matrix(size)
    data = []
    lambda_acc = max(abs(np.linalg.eig(A)[0]))
    for eps in range(-6, -1):
        eps = 10**eps
        data.append({
            'eps': eps,
            'Степенной метод (К-во итераций)':
             power_method(A, eps)[1],
            'Степенной метод |lambda_acc - lambda|' : 
            abs(lambda_acc - abs(power_method(A, eps)[0])),
            'Метод скалярных произведений (К-во итераций)':
            scal_method(A, eps)[1],
            'Метод скалярных произведений |lambda_acc - lambda|' : 
            abs(lambda_acc - abs(scal_method(A, eps)[0]))
        })
    result.append(pd.DataFrame(data))
```
